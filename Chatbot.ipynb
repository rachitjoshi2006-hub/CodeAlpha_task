```python
"""
Project Title : Enhanced UI Chatbot using Python
Created By    : Rachit
Description   :
This project demonstrates a simple rule-based chatbot with an improved
user interface using Gradio Blocks. The chatbot maintains chat history
and provides a clean chat-style experience.

Platform      : Google Colab
Technology    : Python + Gradio
"""

import gradio as gr

# ------------------------------
# Chatbot Logic
# ------------------------------
def chatbot(user_message, chat_history):
    user_message = user_message.lower()

    if "hello" in user_message or "hi" in user_message:
        bot_reply = "Hello! ðŸ˜Š How can I help you?"

    elif "your name" in user_message:
        bot_reply = "I am a chatbot created by Rachit."

    elif "project" in user_message:
        bot_reply = "This is a basic chatbot project with enhanced UI."

    elif "how are you" in user_message:
        bot_reply = "I'm doing great! Thanks for asking ðŸ˜„"

    elif "bye" in user_message:
        bot_reply = "Goodbye! ðŸ‘‹ All the best."

    else:
        bot_reply = "Sorry, I didn't understand that. Try asking something else."

    chat_history.append((user_message, bot_reply))
    return "", chat_history

# ------------------------------
# UI Design using Gradio Blocks
# ------------------------------
with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown(
        """
        # ðŸ¤– Basic Chatbot Project
        ### Created by **Rachit**
        A simple Python chatbot with an improved UI, built entirely in Google Colab.
        """
    )

    chatbot_ui = gr.Chatbot(height=350)
    user_input = gr.Textbox(
        placeholder="Type your message here...",
        show_label=False
    )

    send_btn = gr.Button("Send")

    send_btn.click(
        chatbot,
        inputs=[user_input, chatbot_ui],
        outputs=[user_input, chatbot_ui]
    )

    user_input.submit(
        chatbot,
        inputs=[user_input, chatbot_ui],
        outputs=[user_input, chatbot_ui]
    )

# ------------------------------
# Launch App
# ------------------------------
demo.launch()
```
